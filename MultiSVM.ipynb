```python
#-*- coding:utf-8 -*-'
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
label_cnt = 10
```

```python
data = df.iloc[:,1:].as_matrix()
data = np.insert(data, data.shape[1], 1.0, axis = 1)
label = df.iloc[:,0].as_matrix()
```

```python
w = np.zeros((label_cnt, data.shape[1]))
lr = 1e-8
max_iter_cnt = 4000
iter_cnt = 0
while iter_cnt < max_iter_cnt:
    idx = int(np.random.rand(1)*data.shape[0])
    x = data[idx,:]
    y = label[idx]
    scores = np.dot(w, x)
    scores_correct = scores[y]
    L_i = max(0, np.sum(1-scores_correct+scores))
    if L_i > 0:
        dW = np.zeros_like(w)
        dW += x
        dW[y,:] = (1-label_cnt)*x
        w -= lr*dW
        iter_cnt+=1
#         print('iter %d: loss = %f'%(iter_cnt, L_i))
```

```python
pred = np.dot(w,data.T)
pred = np.argmax(pred, axis=0)
print(np.sum(pred != label)/data.shape[0])
```
